```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("GOOG.csv")
data
```

```python
training_set = data.iloc[:,1:2].values
training_set
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
training_set = sc.fit_transform(training_set)
training_set
```

```python
training_set.size
```

```python
x_train = training_set[0:249]
y_train = training_set[1:250]
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_train = np.reshape(x_train,(249,1,1))
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
regressor = Sequential()
```

```python
regressor.add(LSTM(units=4,activation ='sigmoid',input_shape=(None,1)))
```

```python
regressor.add(Dense(units=1))
```

```python
regressor.compile(optimizer='adam',loss='mean_squared_error')
```

```python
regressor.fit(x_train,y_train,batch_size = 32,epochs = 200)
```

```python
test_set = pd.read_csv("GOOG.csv")
test_set.size
```

```python
real_stock_price = test_set.iloc[:,1:2]
real_stock_price.size
```

```python
inputs = real_stock_price
inputs = sc.fit_transform(inputs)
inputs.size
```

```python
inputs = np.reshape(inputs,(250,1,1))
```

```python
predicted_stock_price = regressor.predict(inputs)
```

```python
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```

```python
import matplotlib.pyplot as plt 
plt.plot(real_stock_price,color='red',label='Real Google Stock Price')
plt.plot(predicted_stock_price,color='blue',label='Predicted Google Stock Price')
plt.title('Google Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Google_stock_price')
plt.legend()
plt.show()
```
